```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

f_juan = np.loadtxt("Juanfile.csv", delimiter=',')
emg_juan = f_juan[:, -1]

time_juan = np.arange(len(emg_juan)) / 1000 # from miliseconds to seconds
plt.plot(time_juan, emg_juan)
plt.plot(time_juan, f_juan[:, 1]*1000) # for the step signal to be noticeable
plt.xlabel('Time in seconds')
plt.ylabel('Voltage in milivolts')

f_bea = np.loadtxt("Csvfile.csv", delimiter=',')
emg_bea = f_bea[:, -1]

plt.figure()
time_bea = np.arange(len(emg_bea)) / 1000
plt.plot(time_bea, emg_bea)
plt.plot(time_bea, f_bea[:, 1]*1000)
plt.xlabel('Time in seconds')
plt.ylabel('Voltage in milivolts')
```
